```python
from cylinder4d import plot_u
import numpy as np
time = 31 # 0 min, 31 max
epoch = 10000
var= "uy"
plane_normal = 'y' # slice plane in x y or z direction, and then plot it
# time, xplane, epoch, velocity/vorticity
err, ref = plot_u(time,  0, epoch, var, plane_normal)
print("abs error : ", err.mean(), err.max())
print("rel error : ", np.median(err/ref), (err/ref).max())
err, ref = plot_u(time, 10, epoch, var, plane_normal)
err, ref = plot_u(time, 20, epoch, var, plane_normal)
err, ref = plot_u(time, 31, epoch, var, plane_normal)
print("abs error : ", err.mean(), err.max())
print("rel error : ", np.median(err/ref), (err/ref).max())
```
